### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
# Exercise 9
```

---

````python
x=np.arange(11)
y1=x**2
y2=x**3
fig, axs = plt.subplots(2, 1, figsize=(8, 10), dpi=80)
axs[0].plot(x, y1, color='g')
axs[0].set_title('y = x^2')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid()
axs[1].plot(x, y2, color='r')
axs[1].set_title('y = x^3')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
plt.show()
````

### Exercise 14:
Plot a filled area under the curve of `y = x^2` for `x` ranging from 0 to 10.

```python
# Exercise 14
```

---

````python
x = np.arange(11)
y = x ** 2  
plt.figure(figsize=(10, 5), dpi=80)
plt.plot(x, y, color='r', label='y = x^2')
plt.fill_between(x, y, color='y', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
````

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.

```python
# Exercise 16
```

---

````python
categories =np.array(['X', 'Y', 'Z'])
group_A = np.array([3, 5, 7])
group_B = np.array([2, 6, 4])
x = np.arange(len(categories))
plt.figure(figsize=(8, 5), dpi=80)
plt.bar(x, group_A, label='Group A', color='b')
plt.bar(x, group_B, bottom=group_A, label='Group B', color='y')
plt.title('Stacked Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, categories)
plt.legend()
plt.show()
````

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
# Exercise 17
```

---

````python
x = np.arange(100)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4
fig, axs = plt.subplots(2, 2, figsize=(10, 10), dpi=80)
axs[0, 0].plot(x, y1, color='b')
axs[0, 0].set_title('y = x')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].grid()
axs[0, 1].plot(x, y2, color='g')
axs[0, 1].set_title('y = x^2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].grid()
axs[1, 0].plot(x, y3, color='orange')
axs[1, 0].set_title('y = x^3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].grid()
axs[1, 1].plot(x, y4, color='r')
axs[1, 1].set_title('y = x^4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].grid()
plt.tight_layout()
plt.show()
````

### Exercise 20:
Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.

```python
# Exercise 20
```

---

````python
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 5), dpi=80)
plt.plot(x, y_sin, label='Sine Wave', color='g', linestyle='--')
plt.plot(x, y_cos, label='Cosine Wave', color='y', linestyle='-')

plt.title('Sine and Cosine Waves')
plt.xlabel('x (radians)')
plt.ylabel('Function value')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
````

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.

```python
# Exercise 21
```

---

````python
x = np.arange(-5, 6)
y = np.arange(-5, 6)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)
plt.figure(figsize=(8, 6), dpi=80)
plt.contour(X, Y, Z, levels=20, cmap='hot')
plt.colorbar()
plt.title('Contour Plot of z = sin(x) + cos(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````
